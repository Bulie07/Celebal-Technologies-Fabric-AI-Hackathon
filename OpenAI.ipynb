#### Function to generate emails for each customer.
This function requires the following parameters
1. CustomerName
2. City
3. Description
4. Price
5. Category
6. AdventureAndSocial
7. ValueAndRole
8. IdeaAndRelaxation

The prompt given to the model is designed to generate an email for a particular customer based on:
1. What shopping behaviour they have shown in the past.
2. What season is going on currently.


```python
def generatePrompt(CustomerName, City, Description, Price, Category, AdventureAndSocial, ValueAndRole, IdeaAndRelaxation):
    
    client = AzureOpenAI(
            api_key = "95a6963d2fbe4959ae63d7bfb154e3fd",
            api_version = "2023-05-15",
            azure_endpoint = "https://activetai.openai.azure.com/"
            )

    userBehaviour = [AdventureAndSocial, ValueAndRole, IdeaAndRelaxation]

    behaviourList = []
    if userBehaviour[0]==1:
        behaviourList.append('Adventure')
        behaviourList.append('Social')
    
    if userBehaviour[1]==1:
        behaviourList.append('Value')
        behaviourList.append('Role')

    if userBehaviour[2]==1:
        behaviourList.append('Idea')
        behaviourList.append('Relaxation')

    # print(behaviourList)
    behaviourString = ', '.join(behaviourList)
    print(behaviourString)
    print()

    assistantRole = f"""You are a digital marketer specializing in email writing for a company called 'Berries'. The company sells winter wear, western wear, Indian wear, and bags. You have access to customer data indicating their past purchases and shopping behaviors. Tailor personalized emails using this data for our customers.

        Training data:
        ---
        {training_data_for_turbo}
        ---
        
        Use the following shopping behavior hints for personalized emails:
        1. Value shopping: Offer discounts with promo codes, vouchers and exclusive deals.
        2. Adventure shopping: Recommend new and exciting products along with emojis.
        3. Idea shopping: Suggest trendy products influenced by market and influencers.
        4. Social shopping: Recommend products for social appreciation.
        5. Role shopping: Suggest products for friends or family.
        6. Relaxation shopping: Recommend products based on their purchase history.

        Personalize the email according to all the behavior customers are showing, provide information such as which trends are going on in the market, provide them relevant discount if needed accordingly. Also, praise them according to the behaviour that they show. Strictly do not include the shopping behaviour names such as Value shopping, Idea shopping and so on.
        
        Compose an email with the following guidelines in mind:

        1. Incorporate humor throughout the email.
        2. Provide recommendations based on the current season as well.
        3. The email should be a minimum of 350 words.
        4. Begin with a lively greeting and conclude with "Warm regards, Team Berries. and an emoji."
        5. Utilize at least 7-8 emojis to enhance the message.
        6. Include the customer's past favourites and new recommendations.
        7. Ensure that the email presents a consolidated message without explicitly separating each shopping behavior. Instead, seamlessly integrate recommendations and personalized content throughout the email, providing a cohesive and engaging experience for the recipient."""


    userRole = f"""Write an email to {CustomerName} residing in {City} who previously bought {Description} from the {Category} at the price of {Price} Rupees. The customer exhibits {behaviourString} shopping behavior."""

    
    completion =  client.chat.completions.create(
            model="turbo",
            messages=[
                {'role': 'system', 'content': assistantRole},
                {'role': 'user', 'content': userRole}
                ],
            temperature=1,
            max_tokens=2000,
            top_p=0.95
        )

    return completion.choices[0].message.content
```

```python
for index, row in salesdata_clean.iterrows():
    print(row.CustomerName, row.City, row.Age, row.Occupation, row.Description, row.Price, row.Category)
    print()
    output = generatePrompt(row.CustomerName, row.City, row.Description, row.Price, row.Category, row.AdventureAndSocial, row.ValueAndRole, row.IdeaAndRelaxation)
    print(output)
    break
```
